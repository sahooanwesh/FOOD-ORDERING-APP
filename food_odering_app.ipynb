```python
import string
import random

class biggy:
    def __init__(self):
        self.foods={}
        self.food=[] 
    
    def create_food(self):
        
        S = 5
        self.food_id=  ''.join(random.choices(string.ascii_uppercase + string.digits, k = S))
        
        n=input(" \n Enter the food Name ")
        self.food.append(n)
        mn=int(input(" \n Enter the quantity "))          
                                                              
        self.food.append(mn)
        p=int(input(" \n Enter the price "))
        self.food.append(p)
        q=int(input(" \n Enter the discount "))
        self.food.append(q)  
        r = int(input('\n Enter stock'))
        self.food.append(r)
        self.foods[self.food_id]=self.food
        
        print(" \n  foods in dict ",self.foods)
        
        self.food = []
        
        
        
        
    def edit_food(self):
            fid=input(" \n Enter your food id ")
            if fid in self.foods:
                a=int(input(" \n  Enter  \n 1 - To change food name \n 2 - To change quantity  \n 3 - To change the food price \n 4 - To change the discount \n 5 - To change the food stock "))
                if a==1:
                    self.foods[fid][0]=input("Please enter new name to the food")
                    print(" \n  Your food's name has been changed sucessfully ")
                if a==2:
                    self.foods[fid][1]=int(input("Please enter new quantity "))
                    print(" \n  Your food's quantity has been changed sucessfully ")
                if a==3:
                    self.foods[fid][2]=int(input("Please enter new price "))
                    print(" \n  Your food's price has been changed sucessfully ")
                if a==4:
                    self.foods[fid][3]=int(input("Please enter new discount "))
                    print(" \n  Your food's discount has been changed sucessfully ")
                if a==5:
                    self.foods[fid][4]=int(input("Please enter new stock "))
                    print(" \n  Your food's stock has been changed sucessfully ")
            
                   
                else:
                    print(" \n  You had entered invalid input please try again ")
            else:
                print(" \n  OOPS !! ...Your food id is not found ")
                
                
                
    def view_food_details(self):                          
        
        fid=input(" \n Enter your food id ")
        if fid in self.foods:
            print(" ")
            print(" \n  Your food id is : ",self.foods[fid])
            print(" Namae of your food :  ",self.foods[fid][0])
            print(" food quantity is : ",self.foods[fid][1])
            print(" food price : ",self.foods[fid][2])
            print(" Current discount : ",self.foods[fid][3])
            print(" Current Stock : ",self.foods[fid][4])
                                   
        else:
            print(" \n  OOPS !! ...Your food id is not found ")
            
    def remove_food(self):             
        fid=input(" \n Enter your food id ")
        if fid in self.foods:
            inp=input(" \n  Do you want to remove this food if Yes type  Y (else) type  N")
            if inp=="Y" or "y":
                      self.foods.pop(fid)
                      print(" \n Your list was updated after removing that food ")
            if inp=="N" or "n":
                    pass
        else:
            print(" \n OOPS !!...Your food id was not found ")
            
            
            
    def menu(self):  
        while True:
                      
            x=int(input(" \n Enter \n 1 - To create food \n 2 - To View food details \n 3 - To update/change food details  \n 4 - To remove food  "))
            
            if x==1:
                self.create_food()
            elif x==2:
                self.view_food_details()
            elif x==3:
                self.edit_food()
            elif x==4:
                self.remove_food()

            else:
                print("You had entered wrong input")
                
                
                
class customer(biggy):
    
    def __init__(self):
        super().__init__()
        self.customers_details={}
        self.order_details={}
        self.sno=1
        
        
    def customer_register(self):
        
        self.cust_name=input("\n Enter your full name ")
        a=int(input("\n Enter your number , it must be 10 digits "))
        if len(str(a))==10:
            self.cust_contact_number=a
        else:
            print("\n Your mobile number does not contains 10 digits ")
        self.mail_id=input("\n Enter your mail-id ")
        self.address=input("\n Enter your address in a single line(i.e without using enter key) ")
        self.password=input("Enter your password : ")
        s={}
        s["customer_name"]=self.cust_name
        s["customer_contact_number"]=self.cust_contact_number
        s["customer_mail_id"]=self.mail_id
        s["customer_address"]=self.address
        s["customer_password"]=self.password
        
        self.customers_details[self.cust_contact_number]=s
        
        print("\n Your registration process completed sucessfully ")
        
        print("\n customers_details \n\n ",self.customers_details)
        
        
        
    def customer_login(self):        
        num=int(input("\n Enter your contact number to login : "))
        if num in self.customers_details:
            pas=input("\n Enter your password")                         
            if self.customers_details[num]["customer_password"]==pas:
                print("\n Logged in sucessfully ")
                f=int(input("\n Enter 1 - New order \n 2 - Order history  \n 3 - update profile"))
                if f==1:
                    k=1
                    for i in self.foods:
                        print("\n The followings are the list that shows current availibility of the foods & their details  ")
                        print("S.no-",k,"\n")                 
                        print("  food ID         : ",i)
                        print("  food name       : ",self.foods[i][0])
                        print("  quantity   : ",self.foods[i][1])
                        print("  Price              : ",self.foods[i][2])
                        print("  Available discount : ",self.foods[i][3])
                        print("  Available stock : ",self.foods[i][4])
                        k+=1
                        
                    sel=input(" \n Enter the food ID of the food that you want to order ")
                    if sel in self.foods:
                        
                        qt=int(input("\n Enter the no.of quantity do you want "))
                        if self.foods[sel][1]>=qt:
                            print("\n The price for ",qt," quantity will be ",qt*(self.foods[sel][2]-self.foods[sel][3])," if ok to continue order type 'y' ")
                            ops=input()
                            if ops=="y" or "Y":
                                di={}
                                d={}
                                self.foods[sel][1]-=qt
                                d["number"]=num
                                d["food ID"]=sel
                                d["food name"]=self.foods[sel][0]
                                d["quantity"]=qt
                
                                self.order_details[self.sno]=d  
                                self.sno+=1
                                print("\n Your order has been created sucessfully ")
                                print(self.order_details)
                        else:
                            print("\n Sorry there is only  ", self.foods[sel][3]," quantity available in stock ")
                    else:
                        print("\n You had entered invalid food id ")
                            
       
                if f==2:
                    for i in self.order_details:
                        if num in self.order_details[i]["number"]:
                            print("\n History of food id - ",self.order_details[i]["food ID"],"  name - ",self.order_details[i]["food name"]," Quantity - ",self.order_details[i]["quantity"])
                    
                       
                        
                    print("-----------")
                    
                
                if f == 3:
                    a=int(input(" \n  Enter  \n 1 - To change name \n 2 - To change contact number \n 3 - To change the email \n 4 - To change the address \n 5 - To change password "))
                    if a==1:
                        self.customers_details[num]["customer_name"]=input("Please enter new name ")
                        print(" \n  Your name has been changed sucessfully ")
                    if a==2:
                        self.customers_details[num]["customer_conatct_number"]=int(input("Please enter new number "))
                        print(" \n  Your number has been changed sucessfully ")
                    if a==3:
                        self.customers_details[num]["customer_mail_id"]=input("Please enter new mail id ")
                        print(" \n  Your mail id has been changed sucessfully ")
                    if a==4:
                        self.customers_details[num]["customer_address"]=input("Please enter new address ")
                        print(" \n  Your address has been changed sucessfully ")
                    if a==5:
                        self.customers_details[num]["customer_password"]=input("Please enter new password ")
                        print(" \n  Your password has been changed sucessfully ")

                
            else:
                print("\n You had entered wrong password")
        else:
            print("\n This number was not registered with us , please verify your number (or) please register first ")
            
    
            
    def menu(self):
        v=True
        
        while v:
            r=int(input("""Enter \n 1 - To create food \n 2 - To View food details \n 3 - To update/change food details  \n 4 - To delet food  \n 5 - For Customer register \n 6 - For Customer login \n 7 - To Exit"""))
            if r==1:
                self.create_food()
            elif r==2:
                self.view_food_details()
            elif r==3:
                self.edit_food()
            elif r==4:
                self.remove_food()
            elif r==5:
                self.customer_register()
            elif r==6:
                self.customer_login()
            elif r==7:
                v=False
        

        else:
            print("You had entered wrong input")
                                                                          
                      
    
        
```
